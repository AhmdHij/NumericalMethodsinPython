```python
import numpy as np
import matplotlib.pyplot as plt
from math import *
def f(x):
    return x*np.cos(x)**2-(x/2)**2

def derivFunc( x ):
    return -np.cos(x)*(2*x*np.sin(x)-np.cos(x)) -x/2
 
# Function to find the root
def newtonRaphson( x ):
    h = f(x) / derivFunc(x)
    while abs(h) >= 0.0001:
        h = f(x)/derivFunc(x)
         
        # x(i+1) = x(i) - f(x) / f'(x)
        x = x - h
     
    print("The value of the root is : ",
                             "%.4f"% x)
 
# Driver program to test above
x0 = -3 # Initial values assumed
newtonRaphson(x0)
```
